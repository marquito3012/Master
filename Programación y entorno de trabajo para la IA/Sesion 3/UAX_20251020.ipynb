```python
%%sql
SELECT *
FROM clientes
WHERE Country = "China"
LIMIT 10;
```

```python
%%sql
SELECT CustomerID AS Identificador,
        Country AS Pais,
        Gender AS Genero
FROM clientes
WHERE Country <> "China"
LIMIT 10;
```

```python
%%sql
SELECT CustomerID AS Identificador,
        Country AS Pais,
        Gender AS Genero
FROM clientes
WHERE CustomerID > 10
And CustomerID <20
LIMIT 10;
```

```python
%%sql
SELECT *
FROM clientes
WHERE Country LIKE 'S%'
LIMIT 5
```

```python
%%sql
SELECT CustomerID AS IDCliente,
       Email AS Correo,
       Country AS Pais,
       Gender AS Genero
FROM clientes
WHERE Country = 'China'
  AND Gender = 'Male'
LIMIT 5
```

```python
%%sql
SELECT Country AS Pais,
       MAX(CustomerID) AS MaxClientes,
       COUNT(CustomerID) AS Clientes,
       MIN(CustomerID) AS MinClientes
FROM clientes
GROUP BY Country
ORDER BY Clientes

```

```python
%%sql
SELECT
  ordenes.OrderID AS Pedido,
  ordenes.CustomerID AS Cliente,
  clientes.Email AS Correo,
  clientes.Country AS Pais,
  clientes.Gender AS Genero
FROM ordenes
LEFT JOIN clientes
  ON ordenes.CustomerID = clientes.CustomerID
WHERE clientes.Country = 'China'
LIMIT 5
```
