```python
%%sql
SELECT clientes.FirstName,
        clientes.LastName,
        productos.Category,
        COUNT(ordenes.OrderID),
        SUM(ordenes.TotalAmount)
FROM clientes
JOIN ordenes ON clientes.CustomerID = ordenes.CustomerID
JOIN productos ON ordenes.OrderID = productos.ProductID
GROUP BY clientes.CustomerID, productos.Category
LIMIT 10;
```

```python
import pandas as pd

# Supongamos que tienes los DataFrames 'clientes', 'ordenes' y 'productos' ya cargados.

# Realizamos los joins entre 'clientes', 'ordenes' y 'productos'
merged_df = pd.merge(clientes[['CustomerID', 'FirstName', 'LastName']],
                     ordenes[['OrderID', 'CustomerID', 'TotalAmount']],
                     on='CustomerID',
                     how='inner')

merged_df = pd.merge(merged_df,
                     productos[['ProductID', 'Category']],
                     left_on='OrderID',
                     right_on='ProductID',
                     how='inner')

# Agrupamos los datos por 'CustomerID' y 'Category' y realizamos las agregaciones necesarias
result = merged_df.groupby(['CustomerID', 'FirstName', 'LastName', 'Category']).agg({
    'OrderID': 'count',       # Contamos el nmero de pedidos
    'TotalAmount': 'sum'      # Sumamos el total de los pedidos
}).reset_index()

# Renombramos las columnas para que coincidan con la consulta SQL
result = result.rename(columns={
    'FirstName': 'FirstName',
    'LastName': 'LastName',
    'Category': 'Category',
    'OrderID': 'OrderCount',       # Cuenta de los pedidos
    'TotalAmount': 'TotalAmountSum' # Suma del total de los pedidos
})

# Mostramos las primeras 10 filas
result.head(10)
```

```python
fechas['day'] = fechas['DateOfBirth'].apply(lambda x: int(x.split('/')[1]))
fechas
```

```python
fechas['month'] = fechas['DateOfBirth'].apply(lambda x: int(x.split('/')[0]))
fechas
```

```python
def quarter(mes):
  if mes in [1,2,3]:
    return 1
  elif mes in [4,5,6]:
    return 2
  elif mes in [7,8,9]:
    return 3
  elif mes in [10,11,12]:
    return 4

fechas['Quarter'] = fechas['month'].apply(quarter)
#fechas.info()
```
