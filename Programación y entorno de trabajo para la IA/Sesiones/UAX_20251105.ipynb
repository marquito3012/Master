```python
%%sql
SELECT clientes.FirstName,
        clientes.LastName,
        productos.Category,
        COUNT(ordenes.OrderID),
        SUM(ordenes.TotalAmount)
FROM clientes
JOIN ordenes ON clientes.CustomerID = ordenes.CustomerID
JOIN productos ON ordenes.OrderID = productos.ProductID
GROUP BY clientes.CustomerID, productos.Category
LIMIT 10;
```

```python
import pandas as pd

# Supongamos que tienes los DataFrames 'clientes', 'ordenes' y 'productos' ya cargados.

# Realizamos los joins entre 'clientes', 'ordenes' y 'productos'
merged_df = pd.merge(clientes[['CustomerID', 'FirstName', 'LastName']],
                     ordenes[['OrderID', 'CustomerID', 'TotalAmount']],
                     on='CustomerID',
                     how='inner')

merged_df = pd.merge(merged_df,
                     productos[['ProductID', 'Category']],
                     left_on='OrderID',
                     right_on='ProductID',
                     how='inner')

# Agrupamos los datos por 'CustomerID' y 'Category' y realizamos las agregaciones necesarias
result = merged_df.groupby(['CustomerID', 'FirstName', 'LastName', 'Category']).agg({
    'OrderID': 'count',       # Contamos el nmero de pedidos
    'TotalAmount': 'sum'      # Sumamos el total de los pedidos
}).reset_index()

# Renombramos las columnas para que coincidan con la consulta SQL
result = result.rename(columns={
    'FirstName': 'FirstName',
    'LastName': 'LastName',
    'Category': 'Category',
    'OrderID': 'OrderCount',       # Cuenta de los pedidos
    'TotalAmount': 'TotalAmountSum' # Suma del total de los pedidos
})

# Mostramos las primeras 10 filas
result.head(10)
```

```python
ordenes['CumulativeShipping'] = ordenes['ShippingFee'].cumsum()
ordenes['CumulativeTotal'] = ordenes['TotalAmount'].cumsum()
ordenes
```

```python
plt.plot(ordenes['OrderDate'], ordenes['CumulativeTotal'], label='Ventas acumuladas')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ordenes['OrderDate'], ordenes['CumulativeTotal'], label='Ventas acumuladas', color = "green")
plt.plot(ordenes['OrderDate'], ordenes['CumulativeShipping'], label='Envio acumulado', color = "purple", linestyle='-.')
plt.xlabel('Fecha de Orden')
plt.ylabel('Total Acumulado')
plt.title('Ventas Acumuladas a lo Largo del Tiempo')
plt.legend()
plt.grid(True)
plt.show()
```

```python
ordenes['YearMonth'] = ordenes['OrderDate'].dt.to_period('M')
ordenes.info()
```

```python
category = productos["Category"].unique()
category

colors = ['green', 'purple','red','blue']
indice = 0
for categoria in category:

    color = colors[indice]
    datos_categoria = productos[productos['Category'] == categoria]
    plt.scatter(datos_categoria['UnitPrice'], datos_categoria['UnitsInStock'], alpha=0.5, label=categoria, color = color)
    indice += 1

plt.xlabel('Precio Unitario')
plt.ylabel('Unidades en Stock')
plt.title('Relacin entre Precio Unitario y Stock por Categora de Producto')
plt.legend(title='Categora')
plt.grid(True)
plt.show()
```

```python
conteo_categorias = productos['Category'].value_counts()
conteo_categorias
```

```python
plt.pie(conteo_categorias, labels=conteo_categorias.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribucin de Productos por Categora')
plt.show()
```

```python
autopct_format = lambda p: f'{p:.1f}% ({int(p * sum(conteo_categorias) / 100)})'

plt.figure(figsize=(8, 8))
colors = ['green', 'blue', 'purple']
plt.pie(conteo_categorias, labels=conteo_categorias.index, autopct=autopct_format, startangle=140, colors = colors)
plt.title('Distribucin de Productos por Categora')
plt.show()
```
