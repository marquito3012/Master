```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
iris = sns.load_dataset('iris')
```

# C√°lculo de coeficientes

```python
media_A = A.mean()
media_B = B.mean()
```

```python
A
```

```python
B
```

```python
list(zip(A,B))
```

```python
num=sum([(x-media_A)*(y-media_B) for (x,y) in list(zip(A,B))])
```

```python
den=sum([(x-media_A)**2 for x in A])
```

```python
beta_1 = num/den
```

```python
beta_1
```

```python
beta_0 = media_B - beta_1*media_A
```

```python
beta_0
```

```python
B_pred = beta_0 + beta_1 * A
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(A, B, color='blue', label='Observaciones')
plt.plot(A, B_pred, color='red', linewidth=2, label='linea regresion')  # L√≠nea de regresi√≥n
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('L√≠nea de Regresi√≥n')
plt.legend()
plt.grid(True)

# Muestra el gr√°fico
plt.show()
```
